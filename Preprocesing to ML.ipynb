```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('clean_data.csv')
df.sample(10)
```

```python
df.info()
```

```python
df.drop(['Code'],axis=1,inplace=True)
df.drop(['Battleground'],axis=1,inplace=True)
# Usuwam te kolumny, gdyż zawiera nr bitwy oraz nazwę, które nic nie wznoszą do uczenia maszynowego.

df['Faction'] = df['Faction'].map({'Horde': 1, 'Alliance': 0})
df['Rol'] = df['Rol'].map({'dps': 1, 'heal': 0})

# Zmieniam na system binarny (0,1)
```

```python
df.to_csv('data_to_ml.csv', index=False)
```
