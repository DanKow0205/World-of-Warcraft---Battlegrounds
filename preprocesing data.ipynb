```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('wowbgs2.csv')
df.sample(10)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
columns = ['Battleground', 'Faction', 'Class', 'Win', 'Lose', 'Rol', 'BE']
for column in columns:
    print(df[column].unique())
```

```python
df['Lose'] = df['Lose'].fillna(0)
df['Win'] = df['Win'].fillna(0)
df['BE'] = df['BE'].fillna(0)
```

```python
df['Lose'] = df['Lose'].astype('int64')
df['Win'] = df['Win'].astype('int64')
df['BE'] = df['BE'].astype('int64')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Class Type
class_type_mapping = {'Warrior': 'Melee', 'Demon Hunter': 'Melee', 'Death Knight': 'Melee', 'Rogue': 'Melee',
                      'Mage': 'Ranged', 'Warlock': 'Ranged', 'Priest': 'Ranged'}
df['Class Type'] = df['Class'].map(class_type_mapping).fillna('Both')

# Pets
pets_mapping = {'Hunter': 1, 'Warlock': 1, 'Death Knight': 1}
df['Pets'] = df['Class'].map(pets_mapping).fillna(0).astype('int64')

# Armor type
armor_type_mapping = {'Mage':'Cloth', 'Warlock':'Cloth', 'Priest':'Cloth', 'Rogue':'Leather', 'Druid':'Leather', 'Monk':'Leather',
                 'Demon Hunter':'Leather', 'Hunter':'Mail', 'Shaman':'Mail'}
df['Armor Type'] = df['Class'].map(armor_type_mapping).fillna('Plate')
```

```python
df.sample(10)
```

```python
df.to_csv('clean_data.csv', index=False)
```
